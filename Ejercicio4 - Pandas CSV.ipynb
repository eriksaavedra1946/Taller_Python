```python
#Línea para mantener los plot dentro del print
%matplotlib inline
#Importar Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table
from IPython.display import display, HTML
plt.close("all")

#variables con listas para realizar filtros de filas
marcas = ['CITROEN', 'FORD', 'KIA', 'HYUNDAI' ]
clases = ['AUTOMOVIL']
#var_2015 = [21600]


#Cargar archivo
data = pd.read_csv('C:/Users/eriks/Documents/Python Scripts/Taller1/guia_fasecolda.csv') #<--- Modificar path depende de donde esté ubicado en su máquina.
print("---------------------------------------------------------------------------------------")
print("Columnas en archivo")
print("---------------------------------------------------------------------------------------")
#imprimir columnas
display(data.columns)
print("---------------------------------------------------------------------------------------")
#Filtro para mostrar solo algunas columnas
df = data[['Marca', 'Clase', 'Referencia1','2015', '2016', '2017', '2018', 'Cilindraje', 'Potencia']]
#Query donde se genera el filtro
print("---------------------------------------------------------------------------------------")
print("Tabla de Valores")
print("---------------------------------------------------------------------------------------")
df2 = df.query("Marca == @marcas and Clase == @clases and Cilindraje != 0" )
display(df2.head())
print("---------------------------------------------------------------------------------------")


#Media de valores de automoviles de 3 marcas excluyendo valores 0
print("---------------------------------------------------------------------------------------")
print("Media de valores de automoviles de 3 marcas excluyendo valores 0")
print("---------------------------------------------------------------------------------------")
df3 = df2[['2015', '2016', '2017', '2018']]
df3 = df3.replace(0, np.NaN)
display(df3.mean(skipna=True))
print("---------------------------------------------------------------------------------------")

#Correlación Potencia x cilindraje x valor
print("---------------------------------------------------------------------------------------")
print("Correlación referencia vehículo x cilindraje x valor")
print("---------------------------------------------------------------------------------------")
df4 = df.query("Marca == @marcas and Clase == @clases and Cilindraje != 0" )
df4 = df2[['Cilindraje', '2018','Potencia']]
display(df4.corr(method ='kendall'))
print("---------------------------------------------------------------------------------------")


#Ranking valores vs referencia autos 2018 para 3 marcas
print("---------------------------------------------------------------------------------------")
print("Ranking valores autos 2018 para 3 marcas")
print("---------------------------------------------------------------------------------------")
df5 = df.query("Marca == @marcas and Clase == @clases and Cilindraje != 0" )
df5 = df2[['Referencia1', '2018']]
df5 = df5.replace(0, np.NaN)
df5['default_rank'] = df5['2018'].rank()
df5['max_rank'] = df5['2018'].rank(method='max')
df5['NA_bottom'] = df5['2018'].rank(na_option='bottom')
df5['pct_rank'] = df5['2018'].rank(pct=True)
display(df5.head(50))
print("---------------------------------------------------------------------------------------")

#Gráfico Correlación Potencia x cilindraje x valor
print("---------------------------------------------------------------------------------------")
print("Gráfico Correlación Potencia x cilindraje x valor")
print("---------------------------------------------------------------------------------------")
df6 = df.query("Marca == @marcas and Clase == @clases and Cilindraje != 0" )
df6 = df2[['Cilindraje', '2018','Potencia']]
plt.matshow(df6.corr(method ='pearson'))
plt.show()
print("---------------------------------------------------------------------------------------")

#Gráfico Cilindraje  vs Potencia
print("---------------------------------------------------------------------------------------")
print("Gráfico Cilindraje  vs Potencia")
print("---------------------------------------------------------------------------------------")
df7 = df.query("Marca == @marcas and Clase == @clases and Cilindraje != 0" )
df7 = df2[['Cilindraje', 'Potencia']]
df7.plot.scatter(x = 'Cilindraje', y = 'Potencia', s = 100);
plt.show()
print("---------------------------------------------------------------------------------------")

#Gráfico comportamiento valor autos 2015 a 2018
print("---------------------------------------------------------------------------------------")
print("Gráfico comportamiento valor autos 2015 a 2018")
print("---------------------------------------------------------------------------------------")
df8 = df.query("Marca == @marcas and Clase == @clases and Cilindraje != 0" )
df8 = df2[['2015', '2016', '2017', '2018']]
df8.plot.box();
```
