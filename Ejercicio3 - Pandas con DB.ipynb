```python
# Importar librerias
import pandas as pd
import sqlite3
from IPython.display import display, HTML

#Objeto de Conexión a SQLite

con = sqlite3.connect("fasecolda.db") # <---- Para correr en maquinas locales, cambiar el path de ubicación de fasecolda.db

#Crea un objeto cursos y se le asigna la conexión
cur = con.cursor()

df = pd.read_sql_query('SELECT [2014], [2015], [2016], [2017], [2018] FROM guia_fasecolda WHERE Clase = "AUTOMOVIL" AND Combustible = "GSL"', con)


print("Total de memoria ram requerida ",df.memory_usage().sum())

#estadistica descriptiva
display(HTML(df.describe().to_html()))
#correlación entre las columnas
display(HTML(df.corr().to_html()))

#Cerrar la conexión
con.close()
```
